# **Estatística Descritiva**

As estatísticas descritivas envolvem os métodos usados para resumir, simplificar e descrever os principais recursos de um conjunto de dados.

Este ramo da estatística visa fornecer uma visão geral concisa e significativa dos dados por meio de medidas como tendência central (média, mediana, moda), dispersão (intervalos, variância, desvio-padrão) e representações gráficas, tais como histogramas, gráficos de barras, gráfico de setores, gráficos de dispersão e etc.

As estatísticas descritivas nos ajudam a compreender as características fundamentais de um conjunto de dados e a torná-lo mais gerenciável para análise.

Infelizmente não temos uma coluna que favoreça a visualização de dados por gráfico de setores, entretanto, vamos fazer mesmo assim para aprender como que fazer e dicas do que NÃO fazer em uma análise de dados com gráfico de setores.

* Dica 1: Se o gráficos está confusos e dificil de entender há algumas possibilidades do que pode estar acontecendo: o gráfico não esta adequado ao tipo de dado ou escolha gráfica está errada.

* Dica 2: Gráfico bom é aquele que batemos o olho e entendemos o que ele quer dizer para nós.

```python
# Grafico de setores
# Erro-Grave: Um grafico de setores só pode ser usado com no maximo 5 categorias.
# O recomendável é até 3 categorias.
plt.figure(figsize=(5,5))
df['palavra_chave'].value_counts().plot.pie()
plt.title('Distribuição de Palavras-Chaves')
plt.ylabel('')
plt.show()
```

```python
# Estilo
# Lembrando: Um grafico de setores só pode ser usado com no maximo 5 categorias.
# O recomendável é até 3 categorias.
# Neste caso seria melhor utilizar uma nuvem de palavras.
plt.figure(figsize=(5, 5))
df['palavra_chave'].value_counts().plot.pie(autopct='%1.1f%%',
                                            startangle=90,
                                            colors=sns.color_palette('pastel'),
                                            wedgeprops=dict(width=0.3))
plt.title('Distribuição de Palavras-chave')
plt.ylabel('')
plt.show()
```

```python
# Agrupando as palavras-chave e somando as contagens
word_counts = df.groupby('palavra_chave').size().sort_values(ascending=True).reset_index(name='count')

# Criando a nuvem de palavras
wordcloud = WordCloud(width=800,
                      height=400,
                      background_color='white').generate_from_frequencies(word_counts.set_index('palavra_chave')['count'])

# Plotando a nuvem de palavras
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='nearest') # 'nearest é recomendado para dados categóricos
plt.axis('off')
plt.show()
```
